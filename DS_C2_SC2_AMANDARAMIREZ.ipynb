**a.Extraer Informaci√≥n del Archivo**

*Cargar previamente el archivo Country Vaccinations en GoogleColab*

```python
import pandas as pd
archivo_country_vaccionations = 'country_vaccinations.csv'
df = pd.read_csv(archivo_country_vaccionations)
df.head()
```

**b. Estructura y Tipos de Datos**

```python
df.info()
df['date'] = pd.to_datetime(df['date'], errors='coerce')
```

**c. Vacunas Aplicadas por Compa√±√≠a**

```python
vacunas_por_compania = df.groupby('vaccines')['total_vaccinations'].sum()

vacunas_por_compania
```

**d. Vacunas aplicadas en todo el mundo**


```python
total_vacunas_mundo = df['total_vaccinations'].sum()

total_vacunas_mundo
```

**e. Promedio de vacunas por pa√≠s**

```python
promedio_vacunas_por_pais = df.groupby('country')['total_vaccinations'].mean()

promedio_vacunas_por_pais
```

**f. Cantidad de vacunas aplicadas el d√≠a 29/01/21 en todo el mundo**

```python
vacunas_29_01_21 = df[df['date'] == '2021-01-29']['daily_vaccinations'].sum()

vacunas_29_01_21
```

**g. Dataframe / Diferencias de aplicaci√≥n de vacunas**


```python
df['diferencias'] = df['daily_vaccinations'] - df['daily_vaccinations_raw']

# Crear un nuevo dataframe con las columnas originales y las diferencias
conDiferencias = df[['country', 'date', 'daily_vaccinations', 'daily_vaccinations_raw', 'diferencias']]

conDiferencias.head()
```

**h. Periodo de tiempo entre el registro con fecha m√°s reciente y registro con fecha m√°s antigua**

```python
fecha_max = df['date'].max()
fecha_min = df['date'].min()

periodo_tiempo = fecha_max - fecha_min

periodo_tiempo
```

**i.Dataframe/ cantidad de vacunas utilizadas cada d√≠a**


```python
df['canVac'] = df['vaccines'].apply(lambda x: len(x.split(',')))

conCantidad = df[['country', 'date', 'vaccines', 'canVac']]

conCantidad.head()
```

**j.Dataframe / Registros anteriores al 20/12/2020**


```python
antes20 = df[df['date'] < '2020-12-20']

antes20.head()
```

**k.Dataframe / vacuna Pfizer**


```python
pfizer = df[df['vaccines'].str.contains('Pfizer', na=False)]
pfizer.head()
```

**l. ResultadosReto**



```python
with pd.ExcelWriter('resultadosReto.xlsx') as writer:
    conDiferencias.to_excel(writer, sheet_name='conDiferencias', index=False)
    conCantidad.to_excel(writer, sheet_name='conCantidad', index=False)
    antes20.to_excel(writer, sheet_name='antes20', index=False)
    pfizer.to_excel(writer, sheet_name='pfizer', index=False)
```
